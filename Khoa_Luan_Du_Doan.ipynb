```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score
from sklearn.metrics import confusion_matrix,accuracy_score,roc_curve,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.utils.validation import column_or_1d
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
```

```python
df = pd.read_csv('du_lieu_chuan.csv')
df
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
X = df.loc[:, df.columns != 'Churn']
y = df.loc[:, df.columns == 'Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state = 0)
```

```python
SScalers = StandardScaler()
X_train = SScalers.fit_transform(X_train)
X_test = SScalers.transform(X_test)
```

```python
model1 = 'Logistic Regression'
```

```python
clf_lr = LogisticRegression()
```

```python
clf_lr.fit(X_train, column_or_1d(y_train))
```

```python
clf_lr.intercept_
```

```python
clf_lr.coef_
```

```python
Yhat_train =clf_lr.predict(X_train)
```

```python
Yhat_test = clf_lr.predict(X_test)
```

```python
clf_lr_conf_matrix = confusion_matrix(y_test, Yhat_test)
```

```python
clf_lr_acc_score = accuracy_score(y_test, Yhat_test)
```

```python
print("confussion matrix")
print(clf_lr_conf_matrix)
print("\n")
print("Accuracy of Logistic Regression:",clf_lr_acc_score*100,'\n')
print(classification_report(y_test,Yhat_test))
```

```python
model2 = 'Support Vector Classifier'
```

```python
clf_svc =  SVC(kernel='rbf', C=2)
```

```python
clf_svc.fit(X_train, y_train)
```

```python
Yhat_svc_train = clf_svc.predict(X_train)
```

```python
Yhat_svc_test = clf_svc.predict(X_test)
```

```python
clf_svc_conf_matrix = confusion_matrix(y_test, Yhat_svc_test)
clf_svc_acc_score = accuracy_score(y_test, Yhat_svc_test)
```

```python
print("confussion matrix")
print(clf_svc_conf_matrix)
print("\n")
print("Accuracy of Support Vector Classifier:",clf_svc_acc_score*100,'\n')
print(classification_report(y_test,Yhat_svc_test))
```

```python
model3 = 'Neural Network'
```

```python
clf_mlp = MLPClassifier(random_state=1, max_iter=300)
```

```python
clf_mlp.fit(X_train, y_train)
```

```python
Yhat_mlp_train = clf_mlp.predict(X_train)
```

```python
Yhat_mlp_test = clf_mlp.predict(X_test)
```

```python
clf_mlp_conf_matrix = confusion_matrix(y_test, Yhat_mlp_test)
clf_mlp_acc_score = accuracy_score(y_test, Yhat_mlp_test)
```

```python
print("confussion matrix")
print(clf_mlp_conf_matrix)
print("\n")
print("Accuracy of Neural Network:",clf_mlp_acc_score*100,'\n')
print(classification_report(y_test,Yhat_mlp_test))
```
