```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = ["Logistic Regression",'SVM','KNN','Decision Tree']
y = [1,30,0.6,0.4]
```

```python
plt.figure(figsize=(10,10))
g=sns.barplot(x = X,y = y)
plt.title('Biểu đồ thời gian huấn luyện ')
g.set_ylabel('Thời gian')
g.set_xlabel('Mô hình')
plt.show()
```

```python
a = ["OnlineSecurity",'OnlineBackup','DeviceProtection',"TechSupport",'StreamingTV','StreamingMovies','InternetService']
b = [18.2,16.4,18.3,16.6,8.5,8.5,5.9]
```

```python
plt.figure(figsize=(10,10))
g=sns.barplot(x = a,y = b)
plt.title('Biểu đồ thời gian dự đoán về lĩnh vực dịch vụ ')
g.set_ylabel('Thời gian')
g.set_xlabel('Dịch vụ')
plt.show()
```
